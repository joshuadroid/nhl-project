```python
# Dependencies
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
import matplotlib.pyplot as plt
```

```python
shots_17_df = pd.read_csv("Resources/shots_2017.csv")
```

```python
# Show the dataframe
shots_17_df.head()
```

```python
shots_16_df = pd.read_csv("Resources/shots_2016.csv")
```

```python
shots_16_df.head()
```

```python
#combine dataframes
frames = [shots_16_df, shots_17_df]

combined_shots_df = pd.concat(frames)

combined_shots_df
```

```python
reduced_df = combined_shots_df[["goal", "xCordAdjusted", "yCordAdjusted", "shotAngleAdjusted", "shotDistance", "shotType",
                                "shooterLeftRight"]].copy()
reduced_df
```

```python
reduced_df["shotType"].fillna("NONE", inplace=True)
reduced_df.isna().sum()
```

```python
X = reduced_df.copy()
y = reduced_df["goal"].copy()
X = X.drop("goal", axis=1)

reduced_df.info()
```

```python
# Use get_dummies to create all numeric values
X = pd.get_dummies(X)
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=13)
```

```python
# Scale the data
scaler = StandardScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# # Train the Logistic Regression model on the scaled data and print the model score
from sklearn.linear_model import LogisticRegression
lrs = LogisticRegression(solver="liblinear")

lrs.fit(X_train_scaled, y_train)
print("Testing Score:", lrs.score(X_test_scaled, y_test))
print("Training Score:", lrs.score(X_train_scaled, y_train))
```

```python
# Create confusion matrix
from sklearn.metrics import confusion_matrix, classification_report

y_true = y_test
y_pred = lrs.predict(X_test_scaled)
confusion_matrix(y_true, y_pred)
```

```python
# confusion matrix accuracy
tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
accuracy = (tp + tn) / (tp + fp + tn + fn) # (111 + 128) / (111 + 5 + 128 + 6)
print(f"Accuracy: {accuracy}")
```

```python
# print predictions
print(f'Actual:\t\t{list(y_test[:100])}')
print(f'Predicted:\t{list(lrs.predict(X_test[:100]))}')
```

```python
# Import dependencies for roc_curve and auc
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
y_pred_proba = lrs.predict_proba(X_test_scaled)[:,1]
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure()
plt.plot(fpr, tpr, color='darkorange',
         label=f'ROC curve (area = {roc_auc:0.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Shot Predictor')
plt.legend(loc="lower right")
plt.show()
```
